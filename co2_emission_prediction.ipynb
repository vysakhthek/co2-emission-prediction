## Masking to create test and train dataset

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
train.count()
```

```python
test.count()
```

## Modeling (Engine Size VS Co2 Emissions)

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
reg.fit(train_x, train_y)

print('Coefficents: ', reg.coef_)
print('Intrecept: ', reg.intercept_)
```

## Plot outputs

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.plot(train_x, reg.intercept_ + reg.coef_*train_x , color='red')
plt.xlabel("Engine Size")
plt.ylabel("Co2 Emissions")
plt.show()
```

## Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
predict_y = reg.predict(test_x)
```

```python
print("Mean Squared Error: ", mean_squared_error(test_y, predict_y))
print("r2 value:", r2_score(test_y, predict_y))
```
